```python
import pandas as pd
import numpy  as np
import scipy  as sp
import chart_studio.plotly as py
import plotly.graph_objects as go
```

```python
z = pd.read_csv("./zbMathScore.csv");z.count()
```

```python
z['pUl']=z['ulftp']/z['p'];
z['pMr']=z['mrtp']/z['p'];
z['pRf']=z['reftp']/z['p'];
z['x']=z['msc'].apply(lambda x: str(x)+"*")
tot=z['p'].sum()
z['rUl']=z['ulftp']/(z['p'])
z['rMr']=z['mrtp']/(z['p'])
z['rRf']=z['reftp']/(z['p'])
z['fUl']=2*(z['rUl']*z['pUl'])/(z['rUl']+z['pUl'])
z['fRf']=2*(z['rRf']*z['pRf'])/(z['rRf']+z['pRf'])
z['fMr']=2*(z['rMr']*z['pMr'])/(z['rMr']+z['pMr'])
z=pd.DataFrame(z[z['p']>200])
z=z.sort_values(by=['fMr'],ascending=False)
z
```

```python
data=[go.Bar(x=z.msc,y=z.fMr,name="MR"),
      go.Bar(x=z.msc,y=z.fRf,name="Ref"),
      go.Bar(x=z.msc,y=z.fUl,name="S14")    
      ];
layout=go.Layout(barmode='group',
                 xaxis={'type':'category','title':'msc', 'mirror': True},
                 yaxis={'title':'f1-score','range':[0,1],'mirror': True},
                 template='simple_white',
                showlegend=False)
go.Figure(data,layout).show()
```
